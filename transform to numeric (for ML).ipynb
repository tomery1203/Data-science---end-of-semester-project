```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv('DataFrame#2.csv',index_col=0)
df2 = df.copy()
```

```python
le = preprocessing.LabelEncoder()
lable = le.fit_transform(df2['developer'])
print(le.classes_)
print (lable)
df2 = df2.drop("developer", axis = 'columns')
df2 ['developer'] = lable
df2
```
