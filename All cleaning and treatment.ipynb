```python
def modify_console_name(df):
  
    ConsoleNames={'PlayStation 3':'PS3',
          'Xbox 360':'X360',
          'Wii':'Wii',
          'Xbox One':'XOne',
          'PC':'PC',
          'Switch':'NS',
          'PlayStation 4':'PS4',
          'Wii U':'WiiU',
          '3DS':'3DS',
          'Xbox Series X':'XS',
          'DS':'DS',
          'PlayStation 5':'PS5',
          'PlayStation Vita':'PSV',
          'PlayStation 2':'PS2'}
    df['platforms'].replace(ConsoleNames,inplace=True)
   
    return df
```

```python
def manipulate_genre(df):
    genre_list=[]
    for i in df['genre']:
        if i is None:
            genre_list.append(np.nan)
        else:
            name=i.split('/')[0]
            if name=='Moder' or name=='2D' or name=='3D':
                name=i.split('/')[1]
                genre_list.append(name)
                continue
            else:
                genre_list.append(name)
    df.drop(columns='genre',inplace=True)
    df.insert(5,'genre',genre_list)
    return df  
```

```python
def normalize_rating(df):
    rating=df['rating'].tolist()
    new_rating=[]
    for i in rating:
        if i ==np.nan:
            print('nan')
            new_rating.append(np.nan)
        else: new_rating.append(i/10)
    df.drop(columns='rating',inplace=True)
    df.insert(7,'rating',new_rating)
    return df
```

```python
def hit_col(df):
    hit=[]
    for i in df['total sales/mil']:
        if i>=1:
            hit.append(1)
        else:
            hit.append(0)
    df.insert(1,'hit',hit)
    return df
    
```

```python
def outliers(df):
    Q1 = np.percentile(df['price'], 25)
    Q3 = np.percentile(df['price'], 75)
    IQR = Q3 - Q1
    df['price'][(df['price'] < Q1 - 1.5*IQR) | (df['price'] > Q3 + 1.5*IQR )] =df['price'].median()
    df['number of players'].replace(np.nan,1,inplace=True)
    df['price'].replace(np.nan,df['price'].mean(),inplace=True)

    return df
```
